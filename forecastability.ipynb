```python
import statistics


def type(df):
   if len(df)>1:
    CV_2=(statistics.stdev(df['products_quantity'])/df['products_quantity'].mean())**2
    ADI=len(df)/len(df[df['products_quantity']!=0])
    if ADI<1.32 and CV_2<0.49:
      return 'smooth'
    elif ADI>=1.32 and CV_2<0.49:
      return 'intermitten'
    elif ADI<1.32 and CV_2>=0.49:
      return 'erratic'
    else:
      return 'lumpy'
```
